## Randomly select + generate random string

```python
mut.shape
```

```python
# number of fake samples
fake_K = 10
```

```python
# random idx
random.seed(0)
random_idx = random.sample(list(np.arange(mut.shape[0])), fake_K)
```

```python
''.join(random.choices(string.ascii_letters + string.digits, k=10))
```

```python
# fake sample IDs
fake_samples = [''.join(random.choices(string.ascii_letters + string.digits, k=10)) for i in range(fake_K)]
```

```python
fake_samples
```

```python
# fake responses
fake_outcomes = [random.sample([0,1], 1)[0] for i in range(fake_K)]
fake_outcomes
```

```python
# fake data
fake_mut = mut.iloc[random_idx].reset_index(drop=True)
fake_cna = cna.iloc[random_idx].reset_index(drop=True)
fake_cnd = cnd.iloc[random_idx].reset_index(drop=True)
fake_cov = cov.iloc[random_idx].reset_index(drop=True)
fake_pdf = pd.DataFrame({'sample':fake_samples, 'outcomes':fake_outcomes})

# deidentify patients
fake_mut['sample'] = fake_samples
fake_cna['sample'] = fake_samples
fake_cnd['sample'] = fake_samples
fake_cov['sample'] = fake_samples
```
